```python
import pandas as pd

import warnings
warnings.filterwarnings("ignore")

from selenium import webdriver
from bs4 import BeautifulSoup
import re
import time

data = pd.read_excel('./data/성북구_행정동별_생활인구수.xlsx')
data_top5 = data.head(5)
keywords = data_top5['행정동명']

for keyword in keywords:
    shop = pd.read_csv('./data/'+keyword+" 음식집 카카오맵 데이터 크롤링.csv")
    shop = shop.sort_values(by = '평가인원', ascending = False)
    target = shop.iloc[0, :]
    
    url = []
    url = target['링크주소']
    columns = ['score', 'review']
    
    excutable_path = './data/chromedriver.exe'
    driver = webdriver.Chrome(executable_path=excutable_path)
    source_url = "https://map.kakao.com/"

    df = pd.DataFrame(columns=columns)
    
    print(keyword, '음식집(카카오맵) 리뷰 데이터 수집 시작')
    print()

    driver.get(url)
    time.sleep(2)

    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    contents = soup.find(name="div", attrs={"class":"evaluation_review"})

    rates = contents.find_all(name="em", attrs={"class":"num_rate"})
    reviews = contents.find_all(name="p", attrs={"class":"txt_comment"})
        
    for rate, review in zip(rates, reviews):
        row = [rate.text[0], review.find(name="span").text]
        series = pd.Series(row, index=df.columns)
        df = df.append(series, ignore_index=True)

    for button_num in range(2, 6):
        try:
            more_reviews = driver.find_element_by_xpath("//a[@data-page='" + str(button_num) + "']")
            more_reviews.click()
            time.sleep(2)

            html = driver.page_source
            soup = BeautifulSoup(html, 'html.parser')
            contents = soup.find(name="div", attrs={"class":"evaluation_review"})

            rates = contents.find_all(name="em", attrs={"class":"num_rate"})
            reviews = contents.find_all(name="p", attrs={"class":"txt_comment"})
                
            for rate, review in zip(rates, reviews):
                row = [rate.text[0], review.find(name="span").text]
                series = pd.Series(row, index=df.columns)
                df = df.append(series, ignore_index=True)
                    
        except ElementNotInteractableException:
            print('not found')
        except:
            break    
                
    df.to_csv("./data/"+keyword+" 음식집 카카오맵 리뷰 데이터 크롤링_단일.csv", index = False, mode='w', encoding = 'utf-8-sig')
    print(keyword+' 완료'+'================')
    print()
    driver.close()
```
