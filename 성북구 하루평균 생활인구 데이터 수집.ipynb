```python
import pandas as pd

df1_raw = pd.read_csv('./data/LOCAL_PEOPLE_DONG_202111.csv', index_col = False)
df2_raw = pd.read_excel('./data/행정동코드_매핑정보_20200325.xlsx')
```

```python
df2_raw
```

```python
df1 = df1_raw.groupby('행정동코드').mean() # 행정동코드별로 하루 평균 생활인구 분류
df1 = df1["총생활인구수"]
df1 = pd.DataFrame(df1)
df1.reset_index(level=0, inplace=True)
df1 = df1.sort_values(by = '행정동코드')
df1
```

```python
# 두번째 데이터셋 전처리
df2 = df2_raw.drop(index = 0, axis=0)
df2 = df2.drop(columns ='통계청행정동코드')
df2 = df2.rename(columns={'행자부행정동코드': '행정동코드'})
df2 = df2.astype({'행정동코드':'int64'})
df2 = df2.sort_values(by = '행정동코드')
df2 = df2.reset_index(drop=True)
df2
```

```python
sbg2 = df2[df2.시군구명 == '성북구']
sbg2
```

```python
sbg1 = df1.loc[df1['행정동코드'].isin(sbg2.행정동코드)]
sbg1
```

```python
result = pd.concat([sbg1, sbg2], axis = 1)
result = result.drop(columns ='행정동코드')
result = result[['시도명', '시군구명', '행정동명', '총생활인구수']]
result = result.sort_values(by = '총생활인구수', ascending = False)
result
```

```python
# xlsx 파일 만들기
result.to_excel(excel_writer='./data/성북구_행정동별_생활인구수.xlsx', index = False)
```
