```python
import pandas as pd
import re

import warnings
warnings.filterwarnings(action='ignore') 

df_ys1_raw = pd.read_csv('./data/길음1동 음식집 카카오맵 리뷰 데이터 크롤링_단일.csv')
df_ys1 = df_ys1_raw

df_ys1 = df_ys1.dropna(axis=0)

def text_purify(text):
    hangul = re.compile('[^ ㄱ-ㅣ가-힣]+') # 한글 정규식 표현
    result = hangul.sub('', text)
    return result

df_ys1['review'] = df_ys1['review'].apply(lambda x: text_purify(x))
ys1_reviews = "".join(df_ys1['review'].tolist())
```

```python
from konlpy.tag import Okt
from collections import Counter

nouns_classify = Okt()
nouns_ys1_reviews = nouns_classify.nouns(ys1_reviews)
count_ys1_reviews = Counter(nouns_ys1_reviews)

stopwords_korean_path = './data/stopwords.txt'
with open(stopwords_korean_path, encoding = 'utf-8') as f:
    stopwords = f.readlines()
stopwords = [x.strip() for x in stopwords]

count_ys1_reviews = Counter({x : count_ys1_reviews[x] for x in count_ys1_reviews if x not in stopwords})

import random
import pytagcloud
import webbrowser

tag_rank_ys1 = count_ys1_reviews.most_common(10)
tag_list_ys1 = pytagcloud.make_tags(tag_rank_ys1, maxsize = 60)

pytagcloud.create_tag_image(tag_list_ys1, './data/길음1동_리뷰_wordcloud.jpg', size=(400,200), fontname = 'Malgun', rectangular = False)

print("상위 10개 키워드 순위")
print(tag_rank_ys1)

from IPython.display import Image
Image(filename = './data/길음1동_리뷰_wordcloud.jpg')
```

```python
import pandas as pd
import re

import warnings
warnings.filterwarnings(action='ignore') 

# 데이터셋 불러오기
df_yu_raw = pd.read_csv('./data/석관동 음식집 카카오맵 리뷰 데이터 크롤링_단일.csv')
df_yu = df_yu_raw

# 리뷰가 없는 결측치 제거
df_yu = df_yu.dropna(axis=0)

# 특수문자 제거 함수
def text_purify(text):
    hangul = re.compile('[^ ㄱ-ㅣ가-힣]+') # 한글 정규식 표현
    result = hangul.sub('', text)
    return result

# 리뷰 한글 정규식 표현으로 정제(특수문자 등 제거)
df_yu['review'] = df_yu['review'].apply(lambda x: text_purify(x))

# 모든 리뷰 하나의 리스트로 합침
yu_reviews = "".join(df_yu['review'].tolist())
```

```python
# 여의동 긍정리뷰 워드 클라우드
from konlpy.tag import Okt
from collections import Counter

# 긍정 리뷰 명사 단위 키워드 추출
nouns_classify = Okt()
nouns_yu_reviews = nouns_classify.nouns(yu_reviews)
count_yu_reviews = Counter(nouns_yu_reviews)

# 긍정 리뷰 키워드 정제 - 불용어(stopword) 불러오기
stopwords_korean_path = './data/stopwords.txt'
with open(stopwords_korean_path, encoding = 'utf-8') as f:
    stopwords = f.readlines()
stopwords = [x.strip() for x in stopwords]

# 긍정 리뷰 키워드 정제 - stopword 제거
count_yu_reviews = Counter({x : count_yu_reviews[x] for x in count_yu_reviews if x not in stopwords})

import random
import pytagcloud
import webbrowser

# 가장 많은 상위 20개 키워드
tag_rank_yu = count_yu_reviews.most_common(20)

# 최대 사이즈 지정
tag_list_yu = pytagcloud.make_tags(tag_rank_yu, maxsize = 60)

# 워드클라우드 이미지 파일 생성
pytagcloud.create_tag_image(tag_list_yu, './data/석관동_리뷰_wordcloud.jpg', size=(400,200), fontname = 'Malgun', rectangular = False)

print("상위 20개 키워드 순위")
print(tag_rank_yu)

# 워드클라우드 이미지 불러오기
from IPython.display import Image
Image(filename = './data/석관동_리뷰_wordcloud.jpg')
```
